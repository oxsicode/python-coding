```python
from bisect import bisect

marks = [100, 70, 50, 35, 0]
grade = ['A', 'B', 'C', '']

def cal_grade(m):
    idx = bisect(marks, m) - 1
    return f'{grade[idx - 1]} Grade.'

stu_mark = int(input('Enter Mark: '))
cal_grade(stu_mark)
```
